```python
x=jio['Ad Revenue Estimate (INR)']
y=jio.Year
plt.title("AD REVENUE")
from matplotlib import style
style.use('ggplot')
plt.plot(x,y,'g',label="Revenue",linewidth=5,marker='o')
plt.legend()
plt.grid(True,color='k')
plt.xlabel('Ad Revenue Estimate (INR)')
plt.ylabel('Year')
plt.show()
```

```python
x=jio['Ad Revenue Estimate (INR)']
y=jio.Year
plt.title("AD REVENUE")
plt.scatter(x,y,label='Revenue',color='k')
plt.xlabel('Ad Revenue Estimate (INR)')
plt.ylabel('Year')
plt.legend()
plt.grid(True,color='y')
autopct='%1.1f%%'
plt.show()
```
