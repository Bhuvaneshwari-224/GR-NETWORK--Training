```python
from bokeh.plotting import figure, show
```

```python
import pandas as p
mov=pd.read_csv(r"C:\Users\hp\Desktop\Movies.csv")
mov
```

```python
from bokeh.plotting import figure, show

p = figure(width=400, height=400)

# add a scatter circle renderer with a size, color, and alpha
p.scatter(mov.Rank, mov['Total Revenue (₹ Cr)'],size=20, color="navy", alpha=0.5)

# show the results
show(p)
```

```python
from bokeh.plotting import figure, show
p = figure(width=400, height=400)

# add a square scatter renderer with a size, color, and alpha
p.scatter(mov.Rank,mov['Budget (₹ Cr)'],marker="square",
          size=20, color="olive", alpha=0.5)

# show the results
show(p)
```

```python

p = figure(width=400, height=400)

# add a line renderer
p.line(mov.Rank,mov['Budget (₹ Cr)'],line_width=2)

show(p)
```

```python

p = figure(width=400, height=400)

# add a line renderer
p.line(mov.Rank,mov['Total Revenue (₹ Cr)'],line_width=2)

show(p)
```

```python
p = figure(width=400, height=400)

# add a line renderer
p.line(mov.Rank,mov['International Revenue (₹ Cr)'],line_width=2)

show(p)
```

```python
p = figure(width=400, height=400)

# add both a line and circles on the same plot
p.line(mov.Rank,mov['International Revenue (₹ Cr)'], line_width=2)
p.scatter(mov.Rank,mov['International Revenue (₹ Cr)'], fill_color="white", size=8)

show(p)
```

```python
p = figure(width=400, height=400)

# add a steps renderer
p.step(mov.Rank,mov['Budget (₹ Cr)'], line_width=2, mode="center")

show(p)
```
