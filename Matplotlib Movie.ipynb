```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

from matplotlib import style
mov=pd.read_csv(r"C:\Users\hp\Desktop\Movies.csv")
mov
#
```

```python

style.use('ggplot')
plt.plot(mov.Rank,mov['Domestic Revenue (₹ Cr)'],'g',label="Revenue",linewidth=3)
plt.legend()
plt.grid(True,color='w')
plt.title("Line Graph")
plt.xlabel("Rank")
plt.ylabel(" Domestic Revenue")
plt.show()
```

```python

style.use('ggplot')
plt.plot(mov.Rank,mov['International Revenue (₹ Cr)'],'y',label="Revenue",linewidth=2)
plt.legend()
plt.grid(True,color='w')
plt.title("Line Graph")
plt.xlabel("Rank")
plt.ylabel(" International Revenue")
plt.show()
```

```python


plt.bar(mov.Rank,mov['International Revenue (₹ Cr)'],label="Revenue",width=0.7)
plt.legend()
plt.title("Bar Graph")
plt.xlabel("Rank")
plt.ylabel("International Revenue")
plt.show()
```

```python
plt.scatter(mov['Rank'],mov['Budget (₹ Cr)'],label='skitscat',color='k')
plt.title('scatter')
plt.legend()
plt.grid(True,color='w')
plt.show()

```

```python

cols=['c','m','r','b']
plt.pie(mov['International Revenue (₹ Cr)'],
        colors=cols,
        startangle=90,
        shadow=True,
        autopct='%1.1f%%'
)
plt.show()
```

```python

style.use('ggplot')
plt.plot(mov['Rank'],mov['Total Revenue (₹ Cr)'],'b',label="Revenue",linewidth=3)
plt.legend()
plt.grid(True,color='w')
plt.title("Line Graph")
plt.xlabel("Rank")
plt.ylabel(" Total Revenue")
plt.show()
```

```python

plt.bar(mov['Movie Title'],mov['Budget (₹ Cr)'],label="Budget",width=0.7,color='c')
plt.legend()
plt.title("Bar Graph")
plt.xlabel("Movie")
plt.ylabel("Budget")
plt.show()
```
