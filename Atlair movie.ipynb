```python
import altair as alt
import pandas as pd
```

```python
mov=pd.read_csv(r"C:\Users\hp\Desktop\Movies.csv")
mov
```

```python
alt.Chart(mov).mark_bar().encode(
    x='Movie Title',  # 'Genre' must be a valid column name
   y='Budget (₹ Cr)',
   color='Rank'
)
```

```python
alt.Chart(mov).mark_point().encode(
    x='Rank',
    y='Total Revenue (₹ Cr)'
)
```

```python
alt.Chart(mov).mark_point().encode(
    x='Movie Title',
    y='Budget (₹ Cr)'
)
```

```python
alt.Chart(mov).mark_point().encode(
    x='Rank',
    y='Budget (₹ Cr)',
     color='Rank'
   
)
```

```python
alt.Chart(mov).mark_point().encode(
    x='Movie Title',
    y='Budget (₹ Cr)',
     color='Movie Title'
   
)
```

```python
alt.Chart(mov).mark_point().encode(
    x='Movie Title',
    y='Language',
     color='Language'
   
)
```

```python
alt.Chart(mov).mark_bar().encode(
    x=alt.X('International Revenue (₹ Cr)', bin=True),
    y='count()'
)
```

```python
alt.Chart(mov).mark_bar().encode(
    x=alt.X('Budget (₹ Cr)', bin=True),
    y='count()'
)
```

```python
alt.Chart(mov).mark_bar().encode(
    x=alt.X('Budget (₹ Cr)', bin=alt.Bin(maxbins=30)),
    y='count()',
    color='Language'
)
```

```python
alt.Chart(mov).mark_line().encode(
    x='Rank',
    y='Budget (₹ Cr)'
)
```

```python
alt.Chart(mov).mark_area().encode(
    x='Rank',
    y='ci0(Budget (₹ Cr))',
    y2='ci1(International Revenue (₹ Cr))'
)
```
