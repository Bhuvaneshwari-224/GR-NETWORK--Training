```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mov = pd.read_csv(r"C:\Users\hp\Desktop\Movies.csv")

```

```python
mov.head()
```

```python
mov.shape
```

```python
#bar plot
res = sns.barplot(x='Rank', y='Budget (₹ Cr)',data=mov)
plt.show()
```

```python
#changing color 
res=sns.barplot(x='Rank',y='Budget (₹ Cr)',data=mov,color='red')
plt.show()
```

```python
#palette colors
res = sns.barplot(x='Rank', y='Budget (₹ Cr)', data=mov, palette='Set2')
plt.show()
```

```python
res=sns.barplot(x='Rank',y='Total Revenue (₹ Cr)',data=mov,palette='rocket')
plt.show()
```

```python
#count plot

sns.countplot(x='Budget (₹ Cr)',data=mov,palette='Set1')

```

```python

sns.countplot(x='Domestic Revenue (₹ Cr)',data=mov,palette='Set2')
```

```python
#horizontal countplot
sns.countplot(y='Runtime',data=mov,palette='rocket')
```

```python

sns.countplot(x='Domestic Revenue (₹ Cr)',hue='Runtime', data=mov, palette='Set1')
```

```python
#distplot
sns.displot(mov.Rank, bins=10,color='g')
```

```python
#scatterplot
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species')
```

```python
#pairplot
sns.pairplot(iris,hue='species',palette='Set1')
```

```python
#lmplot
sns.lmplot(x='petal_length',y='petal_width',data=iris)
```

```python
sns.lmplot(x='petal_length',y='petal_width',hue='species',data=iris,markers=['o','*','^'])
```

```python
#boxplot
sns.boxplot(x='species',y='sepal_width',data=iris,palette='Set2')
```
