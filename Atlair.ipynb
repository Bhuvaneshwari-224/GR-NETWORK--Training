```python
import altair as alt
alt.Chart(df).mark_bar().encode(
    x= "type",
    y="prestige"
)
```

```python
import altair as alt
alt.Chart(df).mark_circle().encode(
    x= "education",
    y="prestige"
)
```

```python
#color
alt.Chart(df).mark_circle().encode(
    x="education",
    y="prestige",
    color='type'
)
```

```python
#color
alt.Chart(df).mark_circle().encode(
    x="education",
    y="prestige",
    color='type',
    size="income"
)
```

```python
#regression
import pandas as pd
import numpy as np
chart=alt.Chart(df).mark_point().encode(
    x='education',
    y='prestige',
    color='type:N',
    tooltip='income'
)
chart+chart.transform_regression('education','prestige').mark_line()


```
